Image manipulation using OpenCV can be seen below

```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import ImageOps
```

```python
image = Image.open('../images/baboon.png')
# lenna = cv2.imread('../images/lenna.jpg')
lenna = Image.open('../images/lenna.jpg')
cat = Image.open('../images/cat.jpg')
```

```python
# First, let's define a helper function to concatenate two images side-by-side. You will not need to understand the code below at this moment, but this function will be used repeatedly in this tutorial to showcase the results.

def get_concat_h(im1, im2):
    #https://note.nkmk.me/en/python-pillow-concat-images/
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
```

Flipping Images

```python
ln_array=np.array(lenna)
width,height,C=ln_array.shape
print('width, height, C =',width,height,C)
```

```python
# Original Image
plt.figure(figsize=(2,2))
plt.imshow(lenna)
plt.show
```

```python
# Let's create a array of the same size and assign zeros to it.
array_flip=np.zeros((width,height,C), dtype=np.uint8)
```

```python
# let's flip the image and assign the first row of pixels of the original array to the new array's last row. repeat!
for i, row in enumerate(ln_array):
    array_flip [width-1-i,:,:]=row
```

```python
plt.figure(figsize=(2,2))
plt.imshow(array_flip)
plt.show
```

```python
# PIL has various method (using -flip method )
im_flip=ImageOps.flip(lenna)
plt.figure(figsize=(2,2))
plt.imshow(im_flip)
plt.show
```

```python
ln_mirror=ImageOps.mirror(lenna)
plt.figure(figsize=(2,2))
plt.imshow(ln_mirror)
plt.show
```

```python
# Transpose the image
ln_tns=lenna.transpose(1)
plt.figure(figsize=(2,2))
plt.imshow(ln_tns)
plt.show
```

Cropping an Image

```python
# Cropping the image from bottom
upper=150
lower=400
crop_top=ln_array[upper:lower,:,]
plt.figure(figsize=(2,2))
plt.imshow(crop_top)
plt.show
```

```python
# Cropping the image sideway
left=150
right=400
crop_horizontal=crop_top[:,left:right,]
plt.figure(figsize=(2,2))
plt.imshow(crop_horizontal)
plt.show
```
